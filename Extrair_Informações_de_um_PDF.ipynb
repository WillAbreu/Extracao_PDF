```python
# Instaladores das bibliotecas necessárias
!pip install pdfplumber
!pip install openpyxl

import pdfplumber
import openpyxl
import time

# Função para extrair e retornar informações específicas de uma página
def extrair_informacoes(page, ws, row):
    # Extrair o texto da página
    texto = page.extract_text()

    # Nomes para busca de informações dentro do PDF
    if "Nome1:" in texto:
        nome_ponto = texto.split("Nome1:")[1].split("\n")[0].strip()
        ws.cell(row=row, column=1, value=nome_ponto)

    if "Nome2:" in texto:
        codigo_ponto = texto.split("Nome2:")[1].split("\n")[0].strip()
        ws.cell(row=row, column=2, value=codigo_ponto)

    if "Nome3:" in texto:
        identificador_medidor = texto.split("Nome3:")[1].split("\n")[0].strip()
        ws.cell(row=row, column=3, value=identificador_medidor)

    if "Nome4:" in texto:
        topologia = texto.split("Nome4:")[1].split("NomeNome")[0].strip() # Caso necessário cortar uma parte da resposta
        ws.cell(row=row, column=4, value=topologia)

    if "Nome5:" in texto:
        agente_conectado = texto.split("Nome5:")[1].split("NomeNome2")[0].strip() # Caso necessário cortar uma parte da resposta
        ws.cell(row=row, column=5, value=agente_conectado)

    if "Nome6:" in texto:
        agente_medicao = texto.split("Nome6:")[1].split("NomeNome3")[0].strip() # Caso necessário cortar uma parte da resposta
        ws.cell(row=row, column=6, value=agente_medicao)

    if "Nome7:" in texto:
        numero_serie = texto.split("Nome7:")[1].split("NomeNome4")[0].strip() # Caso necessário cortar uma parte da resposta
        ws.cell(row=row, column=7, value=numero_serie)

# Caminho do seu arquivo PDF
pdf_file = 'C:/Seu/Caminho/Aqui.pdf'

# Define o limite de páginas desejado
pagina_inicial = 0  # Página inicial
pagina_final = 100  # Página final

# Nome do arquivo Excel (xlsx) de saída
xlsx_file = 'ArquivoComAsExtracoes.xlsx'

# Início do contador de tempo
start_time = time.time()

# Cria um arquivo Excel
wb = openpyxl.Workbook()
ws = wb.active
ws.append(["Nome1", "Nome2", "Nome3", "Nome4", "Nome5", "Nome6", "Nome7"])
row = 2  # Começar na segunda linha (abaixo do cabeçalho)

# Abre o arquivo PDF usando pdfplumber
with pdfplumber.open(pdf_file) as pdf:
    # Itera por todas as páginas do PDF até atingir o número máximo de páginas
    for pagina_num in range(pagina_inicial - 1, min(pagina_final, len(pdf.pages))):
        page = pdf.pages[pagina_num]
        extrair_informacoes(page, ws, row)
        row += 1

        # Calcula o tempo decorrido até agora em minutos e o imprime
        current_time = time.time()
        elapsed_time_minutes = (current_time - start_time) / 60
        print(f"Página {pagina_num + 1}/{min(pagina_final, len(pdf.pages))} - Tempo decorrido: {elapsed_time_minutes:.2f} minutos")

# Salva o arquivo Excel
wb.save(xlsx_file)

# Fim do contador de tempo e cálculo do tempo total gasto em minutos
end_time = time.time()
total_elapsed_time_minutes = (end_time - start_time) / 60

# Imprime o tempo total gasto em minutos
print(f"Tempo total gasto: {total_elapsed_time_minutes:.2f} minutos")
```
